```python
import pandas as pd
import numpy as np
import sklearn.datasets

# set URL  iris dataset 
dataset = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv'

#and store in dataframe df
df = pd.read_csv(dataset)
df.head()
```

```python
# list the data types for each column
df.dtypes
```

```python
# Identify missing data
# Missing data must have an label identifier ( ? for example)
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#discover null values
missing = df.isnull()
missing.head(5)
```

```python
# Correct data format
df.dtypes
```

```python
# to convert 
df[["petal_width"]] = df[["petal_width"]].astype("int")
df.dtypes
df[["petal_width"]] = df[["petal_width"]].astype("float")
df.dtypes
```

```python
# Data Standardization
df['petal_width'] = 100*df["petal_width"]
df.head()
```

```python
#Data Normalization (Normalization is the process of transforming values of several variables 
#into a similar range between 0 and 1 )
df['petal_width'] = df['petal_width']/df['petal_width'].max() 
df['petal_length'] = df['petal_length']/df['petal_length'].max() 
# show the scaled columns
df[["petal_width","petal_length"]].head() 
```

```python
# Binning : group discrete values into segmented groups. 
# There is an example into exercise...
```

```python
#Indicator variable (or dummy variable)
# Transform in numerical values labeled categories ( ex : fuel type gas or diesel)
# This operation is to align with regression methods.
dummy_variable_1 = pd.get_dummies(df["species"])
dummy_variable_1.head()
```

```python
#Merge the result column to orignal dataframe

df = pd.concat([df, dummy_variable_1], axis=1)
df.head()
```

```python
# drop original column 
df.drop('species', axis = 1, inplace=True)
df.head()
```

```python
#save final result
df.to_csv('clean_df.csv')
```
